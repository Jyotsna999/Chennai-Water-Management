```python
reservoirname1 = go.bar(x = [dataset_levels.POONDI.mean()], y = ["Total"], orientation = "h", name = "Poondi", color = "green")
reservoirname2 = go.bar(x = [dataset_levels.REDHILLS.mean()], y = ["Total"], orientation = "h",  name = "Redhills", color = "red")
reservoirname3 = go.bar(x = [dataset_levels.CHOLAVARAM.mean()], y = ["Total"], orientation = "h",  name = "Cholavaram", color = "blue")
reservoirname4 = go.bar(x = [dataset_levels.CHEMBARAMBAKKAM.mean()], y = ["Total"], orientation = "h",  name = "Chembarambakkam", color = "black")
data = [reservoirname1, reservoirname2, reservoirname3, reservoirname4]
layout = go.Layout(barmode = "stack")
fig = go.Figure (data = data, layout = layout)
py.iplot(fig)
```

```python
poondi=dataset_levels[:].POONDI.mean()
chembar=dataset_levels[:].CHEMBARAMBAKKAM.mean()
chola=dataset_levels[:].CHOLAVARAM.mean()
red=dataset_levels[:].REDHILLS.mean()

sizes1=[poondi,chembar,chola,red]
labels=["Poondi","Chembarambakam","Cholavaram","Redhills"]
 
fig1, ax1 = plt.subplots()
ax1.pie(sizes1, labels=labels, autopct='%1.1f%%')
fig = plt.gcf()
ax1.axis('equal')  
plt.title("Water levels In all reservoirs compared")
plt.tight_layout()
plt.show()
```
